```python
#THE ANSWER

num=int(input())

def integer_mirror(n):
    reversed_num = 0

    while n > 0:
        digit = n % 10          # last digit
        reversed_num = reversed_num * 10 + digit
        n = n // 10             # deleting last digit

    return reversed_num


print(integer_mirror(num))



```

Q1
For a d-digit number, the loop executes d times. Each iteration performs one modulo, one multiplication, one addition, and one integer division, resulting in approximately 4d arithmetic operations.



Q2
Since the number of digits d is proportional to log₁₀(n), the time complexity in terms of the input value n is O(log n).

Total operations : 4d
Time complexity : O(d) = O(log n)
