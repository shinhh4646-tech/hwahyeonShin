```python
def is_balanced(s):
    stack = []              # Stack to store opening symbols
    comparisons = 0         # Counter for comparison operations

    # Dictionary to map closing symbols to opening symbols
    matching = {
        ')': '(',
        ']': '[',
        '}': '{'
    }
    # Traverse each character in the input string
    for char in s:

        # If the character is an opening symbol → push to stack
        if char in "([{":
            stack.append(char)

        # If the character is a closing symbol → perform matching
        elif char in ")]}":
            comparisons += 1    #Count comparison attempt

            #If stack is empty → no matching opening symbol
            if not stack:
                return False, comparisons

            # Pop the most recent opening symbol
            top = stack.pop()

            comparisons += 1    # Count symbol matching comparison

            # Check if symbols correspond
            if matching[char] != top:
                return False, comparisons

    # Final check: stack must be empty for balanced string
    return len(stack) == 0, comparisons


# test
inputs = ["{[()]}", "([)]", "", "((())"]

for s in inputs:
    result, comps = is_balanced(s)
    print(f"Input: {s} → Balanced: {result}, Comparisons: {comps}")
```
