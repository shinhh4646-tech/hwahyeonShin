```python
def max_engagement(posts, left, right):
  if left == right:
    return posts[left]
  mid = (left + right) // 2
  left_max = max_engagement(posts, left, mid)
  right_max = max_engagement(posts, mid + 1, right)
  if left_max['engagement_score'] > right_max['engagement_score']:
    return left_max
  else:
    return right_max
```

```python
def sum_engagement(posts, left, right):
  if left == right:
    return posts[left]['engagement_score']
  mid = (left + right) // 2
  left_sum = sum_engagement(posts, left, mid)
  right_sum = sum_engagement(posts, mid + 1, right)
  return left_sum + right_sum
```

```python
def average_engagement(posts, left, right):
  total_sum = sum_engagement(posts, left, right)
  count = right - left + 1
  return total_sum / count
```

```python
def count_above_threshold(posts, left, right, threshold):
  if left == right:
    if posts[left]['engagement_score'] > threshold:
      return 1, [posts[left]['name']]
    else:
      return 0, []
  mid = (left + right) // 2
  left_count, left_list = count_above_threshold(posts, left, mid, threshold)
  right_count, right_list = count_above_threshold(posts, mid + 1, right, threshold)
  total_count = left_count + right_count
  total_list = left_list + right_list
  return total_count, total_list
```

```python
def merge_sort_by_engagement(posts, left, right):
  if left < right:
    mid = (left + right) // 2
    merge_sort_by_engagement(posts, left, mid)
    merge_sort_by_engagement(posts, mid + 1, right)
    merge(posts, left, mid, right)

def merge(posts, left, mid, right):
  i = left
  j = mid + 1
  temp_list = []
  while i <= mid and j <= right:
    if posts[i]['engagement_score'] > posts[j]['engagement_score']:
      temp_list.append(posts[i])
      i += 1
    else:
      temp_list.append(posts[j])
      j += 1
  while i <= mid:
    temp_list.append(posts[i])
    i += 1
  while j <= right:
    temp_list.append(posts[j])
    j += 1
  for k in range(len(temp_list) - 1):
    posts[left + k] = temp_list[k]
```

```python
def find_peak_hour(likes, left, right):
  mid = (left + right) // 2
  if (mid == 0 or likes[mid - 1] <= likes[mid]) and (mid == 23 or likes[mid + 1] <= likes[mid]):
    return mid, likes[mid]
  if mid > 0 and likes[mid - 1] > likes[mid]:
    return find_peak_hour(likes, left, mid - 1)
  else:
    return find_peak_hour(likes, mid + 1, right)
```
