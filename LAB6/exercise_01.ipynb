```python
#1

class SocialGraph:
    def __init__(self, numUsers):
        # V: number of vertices (users), E: number of edges (relationships)
        self.V = numUsers
        self.E = 0

        # 1. Adjacency Matrix: Initialize a 2D list of size numUsers x numUsers with 0s
        self.adjMatrix = [[0 for _ in range(numUsers)] for _ in range(numUsers)]

        # 2. Adjacency List: Array of empty lists to store neighbors for each user
        self.adjList = [[] for _ in range(numUsers)]

    def add_edge(self, u, v):
        """
        Adds a friendship (edge) between user u and user v.
        Updates both the Adjacency Matrix and Adjacency List.
        """
        # Check if the friendship already exists (assuming undirected graph)
        if self.adjMatrix[u][v] == 0:
            # Update Adjacency Matrix
            self.adjMatrix[u][v] = 1
            self.adjMatrix[v][u] = 1

            # Update Adjacency List
            self.adjList[u].append(v)
            self.adjList[v].append(u)

            # Increment edge count
            self.E += 1

    def display(self):
        """Prints the current state of the graph for verification."""
        print(f"Total Users (V): {self.V}, Total Edges (E): {self.E}")

        print("\n--- Adjacency Matrix ---")
        for row in self.adjMatrix:
            print(row)

        print("\n--- Adjacency List ---")
        for i, neighbors in enumerate(self.adjList):
            print(f"User {i}: {neighbors}")

# --- Functional Verification (Proof of Concept) ---
print("--- Verifying SocialGraph Implementation ---")
# Initialize a graph with 4 users
social_net = SocialGraph(4)

# Test 1: Adding friendships
social_net.add_edge(0, 1)
social_net.add_edge(0, 2)
social_net.add_edge(2, 3)

# Test 2: Verify total count
print(f"Verified Edge Count: {social_net.E}") # Should be 3

# Test 3: Display full structure
social_net.display()
```

```python
#2

class SocialGraph:
    def __init__(self, num_users):
        """
        Initializes the graph with a fixed number of users.
        num_users: The total number of nodes (users) in the graph.
        """
        self.V = num_users
        self.E = 0

        # Adjacency Matrix: Best for checking if two specific users are friends O(1)
        self.adj_matrix = [[0 for _ in range(num_users)] for _ in range(num_users)]

        # Adjacency List: Best for iterating over a user's friends O(degree)
        self.adj_list = [[] for _ in range(num_users)]

    def are_friends(self, u, v):
        """
        Checks if user u and user v are friends.
        Time Complexity: O(1)
        """
        return self.adj_matrix[u][v] == 1

    def add_friendship(self, u, v):
        """
        Adds a bidirectional friendship between u and v.
        """
        if not self.are_friends(u, v):
            # Update Matrix
            self.adj_matrix[u][v] = 1
            self.adj_matrix[v][u] = 1

            # Update List
            self.adj_list[u].append(v)
            self.adj_list[v].append(u)

            self.E += 1

    def remove_friendship(self, u, v):
        """
        Removes the friendship between u and v.
        """
        if self.are_friends(u, v):
            # Update Matrix
            self.adj_matrix[u][v] = 0
            self.adj_matrix[v][u] = 0

            # Update List
            self.adj_list[u].remove(v)
            self.adj_list[v].remove(u)

            self.E -= 1

    def get_friends(self, u):
        """
        Returns a list of all friends for user u.
        Time Complexity: O(1) return, O(degree) to use.
        """
        return self.adj_list[u]

    def get_degree(self, u):
        """
        Returns the number of friends user u has.
        """
        return len(self.adj_list[u])

    def get_num_users(self):
        """
        Returns the total number of users in the network.
        """
        return self.V

    def get_num_edges(self):
        """
        Returns the total number of friendships in the network.
        """
        return self.E

```

```python
# Create a social network with 5 users (ID 0 to 4)
social_net = SocialGraph(5)

print("--- Step 1: Initial State ---")
print(f"Total Users: {social_net.get_num_users()}")
print(f"Total Friendships: {social_net.get_num_edges()}")

print("\n--- Step 2: Adding Friendships ---")
# 0 is friends with 1 and 2
social_net.add_friendship(0, 1)
social_net.add_friendship(0, 2)
# 3 is friends with 4
social_net.add_friendship(3, 4)

print(f"Friendships after addition: {social_net.get_num_edges()} (Expected: 3)")
print(f"User 0's friends: {social_net.get_friends(0)} (Expected: [1, 2])")
print(f"User 0's degree: {social_net.get_degree(0)} (Expected: 2)")

print("\n--- Step 3: Checking Connections (are_friends) ---")
print(f"Are 0 and 1 friends? {social_net.are_friends(0, 1)} (Expected: True)")
print(f"Are 0 and 3 friends? {social_net.are_friends(0, 3)} (Expected: False)")

print("\n--- Step 4: Removing a Friendship ---")
social_net.remove_friendship(0, 1)
print(f"After removing (0,1), are they friends? {social_net.are_friends(0, 1)} (Expected: False)")
print(f"New degree of User 0: {social_net.get_degree(0)} (Expected: 1)")
print(f"Total Friendships: {social_net.get_num_edges()} (Expected: 2)")

print("\n--- Step 5: Final Matrix Representation ---")
for row in social_net.adj_matrix:
    print(row)
```

```python
#3
class SocialGraph:
    def __init__(self, num_users):
        # V: number of vertices, E: number of edges
        self.V = num_users
        self.E = 0
        self.adj_matrix = [[0 for _ in range(num_users)] for _ in range(num_users)]
        self.adj_list = [[] for _ in range(num_users)]

    def add_friendship(self, u, v):
        """Adds a bidirectional friendship."""
        if self.adj_matrix[u][v] == 0:
            self.adj_matrix[u][v] = 1
            self.adj_matrix[v][u] = 1
            self.adj_list[u].append(v)
            self.adj_list[v].append(u)
            self.E += 1

    def get_degree(self, u):
        """Returns the number of friends for user u."""
        return len(self.adj_list[u])

    def is_complete_graph(self):
        """
        Checks if the graph is complete (every node connected to every other node).
        Formula: max_edges = V * (V - 1) / 2
        """
        if self.V < 2:
            return True
        max_possible_edges = (self.V * (self.V - 1)) / 2
        return self.E == max_possible_edges

    def graph_density(self):
        """
        Calculates the density of the graph.
        Formula: 2 * |E| / (|V| * (|V| - 1))
        """
        if self.V < 2:
            return 0.0
        return (2 * self.E) / (self.V * (self.V - 1))

    def degree_distribution(self):
        """
        Returns a dictionary mapping each degree to the count of users with that degree.
        Example: {2: 3} means 3 users have exactly 2 friends.
        """
        dist_map = {}
        for i in range(self.V):
            d = self.get_degree(i)
            # Increment count for degree d
            dist_map[d] = dist_map.get(d, 0) + 1
        return dist_map

# --- Functional Verification (Proof) ---
print("--- Social Network Metrics Verification ---")
# Create a network with 4 users
sg = SocialGraph(4)

# Case 1: Sparse Graph
sg.add_friendship(0, 1)
sg.add_friendship(2, 3)

print(f"1. Is Complete? {sg.is_complete_graph()} (Expected: False)")
print(f"2. Graph Density: {sg.graph_density():.2f} (Expected: 2/6 = 0.33)")
print(f"3. Degree Distribution: {sg.degree_distribution()} (Expected: {{1: 4}})")

# Case 2: Making it a Complete Graph
sg.add_friendship(0, 2)
sg.add_friendship(0, 3)
sg.add_friendship(1, 2)
sg.add_friendship(1, 3)

print("\n--- After Adding More Edges ---")
print(f"1. Is Complete? {sg.is_complete_graph()} (Expected: True)")
print(f"2. Graph Density: {sg.graph_density():.2f} (Expected: 1.00)")
print(f"3. Degree Distribution: {sg.degree_distribution()} (Expected: {{3: 4}})")

```
