```python
import copy

class TrendingHeap:
    def __init__(self):
        self.heap = []
        self.entry_map = {}

    def _swap(self, i, j):
        # Internal Helper: Swaps elements and updates the index map
        self.heap[i], self.heap[j] = self.heap[j], self.heap[i]
        self.entry_map[self.heap[i][1]] = i
        self.entry_map[self.heap[j][1]] = j

    def _bubble_up(self, idx):
        # Internal Helper: Moves an element up to maintain Max-Heap order
        while idx > 0:
            parent = (idx - 1) // 2
            if self.heap[idx][0] > self.heap[parent][0]:
                self._swap(idx, parent)
                idx = parent
            else:
                break

    def _bubble_down(self, idx):
        # Internal Helper: Moves an element down to maintain Max-Heap order
        length = len(self.heap)
        while True:
            left = 2 * idx + 1
            right = 2 * idx + 2
            largest = idx

            if left < length and self.heap[left][0] > self.heap[largest][0]:
                largest = left
            if right < length and self.heap[right][0] > self.heap[largest][0]:
                largest = right

            if largest != idx:
                self._swap(idx, largest)
                idx = largest
            else:
                break

    # 1. Push
    def push(self, post_id, likes, timestamp):
        new_entry = [likes, post_id, timestamp]
        self.heap.append(new_entry)
        idx = len(self.heap) - 1
        self.entry_map[post_id] = idx
        self._bubble_up(idx)

    # 2. Pop Max
    def pop_max(self):
        if len(self.heap) == 0: return None
        max_post = self.heap[0]
        self._swap(0, len(self.heap) - 1)
        if max_post[1] in self.entry_map:
            del self.entry_map[max_post[1]]
        self.heap.pop()
        if len(self.heap) > 0:
            self._bubble_down(0)
        return max_post

    # 3. Peek Max
    def peek_max(self):
        return self.heap[0] if self.heap else None

    # 4. Get Top K
    def get_top_k(self, k):
        temp_heap = copy.deepcopy(self)
        top_list = []
        for _ in range(min(k, len(temp_heap.heap))):
            top_list.append(temp_heap.pop_max())
        return top_list

    # 5. Update Likes
    def update_likes(self, post_id, new_likes, timestamp):
        if post_id not in self.entry_map:
            return
        idx = self.entry_map[post_id]
        old_likes = self.heap[idx][0]
        self.heap[idx][0], self.heap[idx][2] = new_likes, timestamp
        if new_likes > old_likes:
            self._bubble_up(idx)
        else:
            self._bubble_down(idx)

    # 6. Size
    def size(self):
        return len(self.heap)

# ==========================================
# Simulation Execution
# ==========================================
if __name__ == "__main__":
    th = TrendingHeap()

    print("--- [Simulation: Start] ---")
    th.push("Post_A", 100, 1000)
    th.push("Post_B", 500, 1001)
    th.push("Post_C", 300, 1002)

    print(f"Initial Size: {th.size()}")
    print(f"Top Post: {th.peek_max()}")

    print("\nUpdating Post_A likes to 800...")
    th.update_likes("Post_A", 800, 1010)
    print(f"New Top Post: {th.peek_max()}")

    print("\nGetting Top 2 Ranking:")
    print(th.get_top_k(2))

    print(f"\nFinal Size after pops in Top K (should still be 3): {th.size()}")
```

**_swap(i, j)**: Swaps two elements in the list. Crucially, it also updates the entry_map so the post_id always points to the correct new index. Without this, update_likes would fail.

**_bubble_up(idx)**: When a post gets more likes (or is newly added), it "bubbles up" the tree by swapping with its parent until the Max-Heap property is restored.

**_bubble_down(idx)**: When the top post is removed or a post loses likes, the element at the top (or current index) "bubbles down" by swapping with its largest child until it reaches its correct position.

```python
#3

import math

class TrendingHeap:
    def __init__(self):
        self.heap = []  # Internal array representing the heap
        self.entry_map = {}

    # --- Heap Analysis Functions ---

    def is_valid_heap(self):
        """
        Validates the Max-Heap property.
        Checks if every parent node is greater than or equal to its children.
        """
        # We only need to check nodes that have children (up to length/2 - 1)
        for i in range(len(self.heap) // 2):
            left = 2 * i + 1
            right = 2 * i + 2

            # Check if left child exists and violates Max-Heap property
            if left < len(self.heap) and self.heap[i][0] < self.heap[left][0]:
                return False

            # Check if right child exists and violates Max-Heap property
            if right < len(self.heap) and self.heap[i][0] < self.heap[right][0]:
                return False

        return True

    def get_height(self):
        """
        Returns the height of the heap.
        Calculated as floor(log2(N)) for a complete binary tree.
        """
        if not self.heap:
            return 0
        return math.floor(math.log2(len(self.heap)))

    def get_level_order(self):
        """
        Returns the internal array.
        In a binary heap, the array itself is stored in level-order.
        """
        return self.heap

    # --- Essential methods for the simulation to run ---
    def _swap(self, i, j):
        self.heap[i], self.heap[j] = self.heap[j], self.heap[i]
        self.entry_map[self.heap[i][1]], self.entry_map[self.heap[j][1]] = i, j

    def push(self, post_id, likes):
        self.heap.append([likes, post_id])
        idx = len(self.heap) - 1
        self.entry_map[post_id] = idx
        # Bubble up logic
        while idx > 0:
            p = (idx - 1) // 2
            if self.heap[idx][0] > self.heap[p][0]:
                self._swap(idx, p)
                idx = p
            else: break
```

```python
import random
import time
import copy

class TrendingHeap:
    def __init__(self):
        self.heap = []
        self.entry_map = {}

    def _swap(self, i, j):
        self.heap[i], self.heap[j] = self.heap[j], self.heap[i]
        self.entry_map[self.heap[i][1]] = i
        self.entry_map[self.heap[j][1]] = j

    def _bubble_up(self, idx):
        while idx > 0:
            parent = (idx - 1) // 2
            if self.heap[idx][0] > self.heap[parent][0]:
                self._swap(idx, parent)
                idx = parent
            else: break

    def _bubble_down(self, idx):
        n = len(self.heap)
        while True:
            l, r, largest = 2*idx + 1, 2*idx + 2, idx
            if l < n and self.heap[l][0] > self.heap[largest][0]: largest = l
            if r < n and self.heap[r][0] > self.heap[largest][0]: largest = r
            if largest != idx:
                self._swap(idx, largest)
                idx = largest
            else: break

    def push(self, post_id, likes, timestamp):
        new_entry = [likes, post_id, timestamp]
        self.heap.append(new_entry)
        idx = len(self.heap) - 1
        self.entry_map[post_id] = idx
        self._bubble_up(idx)

    def update_likes(self, post_id, new_likes, timestamp):
        if post_id not in self.entry_map: return
        idx = self.entry_map[post_id]
        old_likes = self.heap[idx][0]
        self.heap[idx][0], self.heap[idx][2] = new_likes, timestamp
        if new_likes > old_likes: self._bubble_up(idx)
        else: self._bubble_down(idx)

    def get_current_likes(self, post_id):
        """Helper: Returns the current likes of a specific post"""
        if post_id in self.entry_map:
            return self.heap[self.entry_map[post_id]][0]
        return 0

    def pop_max(self):
        if not self.heap: return None
        max_val = self.heap[0]
        self._swap(0, len(self.heap)-1)
        del self.entry_map[max_val[1]]
        self.heap.pop()
        if self.heap: self._bubble_down(0)
        return max_val

    def get_top_k(self, k):
        temp_heap = copy.deepcopy(self)
        return [temp_heap.pop_max() for _ in range(min(k, len(self.heap)))]

# --- Main Simulation Function ---

def simulate_trending_feed():
    trending_app = TrendingHeap()

    # 1. Start with 100 random posts
    print("Initializing simulation with 100 random posts...")
    for i in range(1, 101):
        trending_app.push(post_id=i, likes=random.randint(0, 1000), timestamp=time.time())

    start_time = time.perf_counter() # Start measuring total latency

    # 2. Performing 10,000 updates
    print(f"Executing 10,000 updates...")
    for update_count in range(1, 10001):
        target_id = random.randint(1, 100)
        # new_likes = current + random(1, 10)
        new_likes = trending_app.get_current_likes(target_id) + random.randint(1, 10)
        trending_app.update_likes(target_id, new_likes, time.time())

        # 3. Every 1000 updates, display only top 5
        if update_count % 1000 == 0:
            top_5 = trending_app.get_top_k(5)
            # Simplify view: (Likes, PostID)
            display_data = [(p[0], f"ID:{p[1]}") for p in top_5]
            print(f"  > Step {update_count}: {display_data}")

    end_time = time.perf_counter()
    measured_time = end_time - start_time

    # 4. Measure average time per operation
    # Average Latency = Total Duration / Number of Updates
    avg_latency = measured_time / 10000

    print("-" * 60)
    print(f"Simulation Complete.")
    print(f"Total Time: {measured_time:.4f} seconds")
    print(f"Average Latency per Update: {avg_latency:.8f} seconds")

if __name__ == "__main__":
    simulate_trending_feed()
```
