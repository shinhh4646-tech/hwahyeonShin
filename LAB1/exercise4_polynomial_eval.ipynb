```python
def polynomial_evaluation(coef, x):
  n = len(coef) - 1 # Degree

  # if coef empty
  if n < 0:
    return 0

  result = coef[n] # leading coefficient

  for i in range(n-1, -1, -1): # Honer's method
    result = result * x + coef[i]

  return(result)
```

```python
coef1 = [1, 2, 3, 4]
x1 = 2
print(polynomial_evaluation(coef1, x))
```

```python
coef2 = [2]
x2 = 1
print(polynomial_evaluation(coef2, x2))
```

```python
coef3 = []
x3 = 3
print(polynomial_evaluation(coef3, x3))
```
