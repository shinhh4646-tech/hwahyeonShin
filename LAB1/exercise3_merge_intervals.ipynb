```python
def merge(interval):
  interval.sort() # Sort intervals

  # if empty
  if len(interval) == 0:
    return []

  # if not empty
  result = [] # create empty list
  result.append(interval[0]) # first interval in result

  for i in range(1, len(interval)):

    current_interval = interval[i]
    start = current_interval[0]
    end = current_interval[1]

    last_add_interval = result[-1]
    last_add_end = result[-1][1]

    if start <= last_add_end: # if overlap
      result[-1][1] = max(last_add_end, end)

    else: # if not overlap
      result.append(current_interval)

  return result

```

```python
input1 = [[1,3], [2,6], [15,18], [8,10]]
print(merge(input1))
```

```python
input2 = [[1,4], [0,4]]
print(merge(input2))
```
