```python
a = input()

upper_count = 0
punc_count = 0
total_letters = 0
prev_char = None
repeated = 1
spam = False

for i in a:
    if i.isupper():
        upper_count += 1

    elif i == '!' or i == '?':
        punc_count += 1

    total_letters += 1

    if i == prev_char:
      repeated += 1
    else:
      repeated = 1

    prev_char = i

    if repeated >= 3:
      spam = True


caps_ratio = upper_count / total_letters

print(f'Uppercase: {upper_count}, Punctuation: {punc_count}, Caps ratio: {caps_ratio: .2}')

if spam:
  print('Potential spam indicator')
elif caps_ratio >= 0.6 or punc_count >= 5:
  print('AGGRESSIVE')
elif caps_ratio >= 0.3 or punc_count >= 3:
  print('URGENT')
else:
  print('CALM')
```
