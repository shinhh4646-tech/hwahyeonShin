```python
#1. Mutual friends (intersection) – people both users know

def get_mutual_friends(user_a_friends, user_b_friends):
    mutual = user_a_friends & user_b_friends
    return mutual
```

```python
#2. Unique friends of each user (difference)


def print_unique_friends(user_a_friends, user_b_friends):
    unique_a = user_a_friends - user_b_friends
    unique_b = user_b_friends - user_a_friends

    print(f"Unique friends of A: {unique_a}")
    print(f"Unique friends of B: {unique_b}")
```

```python
#3. Suggested friends – friends of friends not already


def get_suggested_friends(target_user, network_map):
    suggestions = set()
    my_friends = network_map.get(target_user, set())

    for friend in my_friends:
        fof_list = network_map.get(friend, set())
        for candidate in fof_list:
            if candidate != target_user and candidate not in my_friends:
                suggestions.add(candidate)

    return suggestions
```

```python
#Jaccard Similarity

def calculate_jaccard_similarity(set_a, set_b):
    intersection = set_a & set_b
    union = set_a | set_b

    if len(union) == 0:
        return 0.0

    return len(intersection) / len(union)
```

```python
#EXAMPLE

user_a_friends = {101, 102, 103, 104, 105}
user_b_friends = {103, 104, 106, 107, 108}

print(f"Mutual: {get_mutual_friends(user_a_friends, user_b_friends)}")
print_unique_friends(user_a_friends, user_b_friends)
print(f"Similarity: {calculate_jaccard_similarity(user_a_friends, user_b_friends)}")
```
